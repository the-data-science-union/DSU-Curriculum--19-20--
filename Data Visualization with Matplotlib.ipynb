```python
##Data Visualization
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn
import pandas as pd

Cars = pd.read_csv('cars04.csv')
Cars.head()
```

```python
#to plot data without having knowledge of what plot to use, then use the plot method
#always add plt.show() to the end of the plot
plt.plot(Cars["CityMPG"])
plt.show()
```

```python
###Histograms
plt.hist(x=Cars["Width"], bins='auto')
plt.xlabel('Width Values') #x-label
plt.ylabel('Frequency') #y-label
plt.title('My Very Own Histogram') #title
plt.show()
```

```python
#Scatterplot
plt.scatter(Cars["EngineSize"],Cars["CityMPG"])
plt.xlabel("Engine Size")
plt.ylabel("City Mileage")
plt.title("My Scatterplot")
plt.show()
```

```python
###Boxplots
###Can be used to solve outliers in your data with re[ect to a single column.
###Best tool to point out outliers alongside leverage points.

plt.boxplot(Cars["EngineSize"])
plt.show()
```

```python
#colors markers, line styles
#simply add the color, linestyle and marker agrguement to add depth to your plot
plt.plot(Cars["CityMPG"],color = 'k')
plt.show()
```

```python
plt.plot(Cars["CityMPG"], linestyle='dashed')
plt.show()
```

```python
plt.plot(Cars["CityMPG"], marker='o')
plt.show()
```

```python
#if your raw you do it all at the sametime
plt.plot(Cars["CityMPG"],'ko--') #the order is color,marker,then linestyle
plt.show()
```

```python
##A figure is a "frame" that stores many different types of plots
#You can add plots to the figure, called subplots if you wish to draw more than 1 plot 

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
```

```python
#The 1st arguement indictaes the x coordinate, the 2nd the y cooridnate, and the last being the speicifc # of the graph.
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(np.arange(40), np.arange(40))
ax2 = fig.add_subplot(2,2,2)
ax2.hist(np.arange(100),bins=20,color='k',alpha=0.3)
ax3 = fig.add_subplot(2,2,3)
ax3.plot(randn(30).cumsum(),'k--')
plt.show()
```

```python
#**Need to define everything at the same time for the plot to show up

#lets say you wanted to add multiple lines to one plot
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.plot(randn(1000).cumsum(),'k',label='one')
ax1.plot(randn(1000).cumsum(),'k--',label='two')
ax1.plot(randn(1000).cumsum(),'k.',label='three')
ax1.legend(loc='best') #add legend
plt.show()
```
